```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import imutils
```

```python
img = cv2.imread('../data/raw/ball.png')
plt.imshow(img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gray_blur = cv2.GaussianBlur(gray, (5,5), 0)
edges = cv2.Canny(gray_blur, 50, 150, apertureSize=3)
plt.imshow(edges, cmap='gray')

sobelx = cv2.Sobel(gray_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
sobely = cv2.Sobel(gray_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
plt.imshow(sobelx, cmap='gray')
```

```python
gray_blur2 = gray_blur.copy()
gray_blur2 = cv2.threshold(gray_blur2, 170, 255, cv2.THRESH_BINARY)[1]
plt.imshow(gray_blur2, cmap='gray')
```

```python
cnts = cv2.findContours(gray_blur2.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
clone = gray.copy()

print(f'{len(cnts)=}')
for (i, c) in enumerate(cnts):
    cv2.drawContours(clone, [c], -1, (0, 255, 0), 2)
    plt.imshow(clone, cmap='gray')
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
from pathlib import Path

def find_contours(img: np.array) -> np.ndarray:
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  gray_blur = cv2.GaussianBlur(gray, (5,5), 250)
  
  gray_blur2 = cv2.threshold(gray_blur.copy(), 250, 255, cv2.THRESH_BINARY)[1]
  plt.imshow(gray_blur2, cmap='gray')

  cnts = cv2.findContours(gray_blur2.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
  cnts = imutils.grab_contours(cnts)
  return cnts
```
